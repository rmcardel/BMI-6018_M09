```python
#Question 4 What proportion of flights to SEA come from the two NYC airports (LGA and JFK)?  float

SEA = flights_data.loc[flights_data['dest']=='SEA']
SEAtotal = len(SEA.index)

#NYSEA = SEA.loc[(SEA['origin']=='JFK') | (SEA['origin']=='LGA')]
# or
import re
NYSEA = SEA.loc[SEA['origin'].str.contains('JFK|LGA', flags=re.I, regex=True)]
NYtotal = len(NYSEA.index)

q4 = NYtotal/SEAtotal
print('The proportion of flights to SEA that come from the two NYC airports is', q4)
```

```python
#Question 5 Which date has the largest average depature delay? Pd slice with date and float
#please make date a column. Preferred format is 2013/1/1 (y/m/d)
# source: https://www.moonbooks.org/Articles/How-to-create-a-datetime-column-from-year-month-and-day-columns-in-pandas-/

df5 = flights_data
df5['Date'] = pd.to_datetime(flights_data[['year', 'month', 'day']])

df5.groupby(['Date']).mean().sort_values('dep_delay', ascending = False)

q5 = '2013-03-08 is the date with the largest average departure delay (83.537).'
print(q5)
```

```python
#Question 6 Which date has the largest average arrival delay? pd slice with date and float

df5.groupby(['Date']).mean().sort_values('arr_delay', ascending = False)

q6 = '2013-03-08 is the date with the largest average arrival delay (85.862)'
print(q6)
```

```python
#%% Numpy Data Filtering/Sorting Question Answering
#Use weather_data_np

import numpy as np
import pandas as pd

weather_data_pd = pd.read_csv('weather.csv')
weather_data_np = weather_data_pd.to_numpy()

arr = weather_data_np
arr
```

```python
#Question 9 How many observations were made in Feburary? Int
# source: https://stackoverflow.com/questions/4455076/how-do-i-access-the-ith-column-of-a-numpy-multidimensional-array
# source: https://note.nkmk.me/en/python-numpy-count/

month = arr[:,3]

q9 = np.count_nonzero(month==2.0)
print(q9,'observations were made in February')
```

```python
#Question 10 What was the mean for humidity in February? Float

# filter to only Feb observations (col 3 = 2.0)
feb = arr[arr[:,3]==2]

# filter Feb observations down to humidity column only (col 8)
humidity = feb[:,8]
q10 = np.mean(humidity)
print('The mean humidity in February was',q10)
```

```python
#Question 11 What was the std for humidity in February? Float
q11 = np.std(humidity)
print('The standard deviation for humidity in February was',q11)
```
